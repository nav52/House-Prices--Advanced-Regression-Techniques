## Feature selection

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

## for feature slection

from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

pd.pandas.set_option('display.max_columns', None)
```

There is a mismatch of the shape of the data with respect to the amount of features in Test and train data after feature engineering.
Let's do the feature selection to select the influential features so that shape can be uniformed for modelling

```python
housing_data = pd.read_csv('X_train.csv')
housing_data.head()
```

```python
# Get dependent feature and drop it from dataset
y_train = housing_data[['SalePrice']]

X_train = housing_data.drop(['Id','SalePrice'], axis=1)
```

```python
# Apply feature selection
# Specify the Lasso Regression model
# Select appropriate alpha (bigger alpha --> less features will be selected)

# Use SelectFromModel that selects the features with non-zero coefficients.

select_features_model = SelectFromModel(Lasso(alpha=0.005, random_state=0))
select_features_model.fit(X_train, y_train)
```

```python
select_features_model.get_support()
# True indicates the important feature and False indicates ignorable features
```

```python
# Let's see the selected features

features_selected = X_train.columns[(select_features_model.get_support())]

print('Total features in the dataset: {}'.format((X_train.shape[1])))
print('Selected features from dataset : {}'.format(len(features_selected)))
print('Features whose Coefs shrank to zero: {}'.format(np.sum(select_features_model.estimator_.coef_ == 0)))
```

```python
features_selected
```

```python
# Let's save the training data with selected features

X_train = X_train[features_selected]
X_train.head()
```

```python
# Let us get the same features for the test data as well.

housing_data_test = pd.read_csv('X_test.csv')

X_test = housing_data_test[features_selected]
X_test.head()
```

```python
# Let's save X_train with the y_train data to use it in modelling phase

X_train = pd.concat([X_train, y_train], axis=1)
X_train
```

```python
X_train.to_csv('X_train_selected.csv', index=False)
X_test.to_csv('X_test_selected.csv', index=False)
```
